2). Average unit price for each region. Use the groupby() method.

```python
df.groupby(['Sepal.Length']).mean()
```

3). Average unit price for each region. Use the pivot_table() method.

```python
table = pd.pivot_table(df,values='Sepal.Length', index=['Sepal.Width','Petal.Length','Petal.Width','Amount'], aggfunc=np.mean)
table
```

6). Total units for each region and item type in one code sentence. Use the pivot_table() method. Fill the missing values with 0.

```python
table = pd.pivot_table(df,values='Sepal.Length', index=['Sepal.Width','Petal.Length','Petal.Width','Amount'], aggfunc=np.sum,fill_value=0)
table
```
